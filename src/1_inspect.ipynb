```python
import json # Used to read .json files , convert json to dict/lists.
import os # Used to check if file exists, and to get the path of the current script.
```

Extracting Diff Clause .

ðŸ“œ What Is a Clause?

A clause is a specific section of a legal contract that talks about one particular rule, obligation, or condition.

```python
def inspect_cuad_data(file_path): # Defines a function named inspect_cuad_data
    print(f"Loading data from {file_path}...")

    if not os.path.exists(file_path): # Checks if the file exists at the specified path
        print(f"Error: File not found at {file_path}")
        return
    else:
        print(f"File found at {file_path}.")

    with open(file_path, 'r') as f: # Opens the file in read mode
        data = json.load(f) # Loads the JSON data into a Python dictionary
        print(f"Data loaded. Top-level keys: {list(data.keys())}")
    
    unique_clause_types = set() # creating an empty set to store unique clause types
    if 'data' in data: #Checks if JSON contains a key called "data"
        documents = data['data'] # Accesses the list of documents under the "data" key
        print(f"Number of documents: {len(documents)}")

        for doc in documents: # now iterate over each contract
            title = doc.get('title', 'Unknown') # Get the title of the contract, or 'Unknown' if not available

            for para in doc.get('paragraphs', []): #If 'paragraphs' missing â†’ default to empty list
                
                for qa in para.get('qas', []): # now iterate over each ques/ans in the paragraph
                    qa_id = qa.get('id', '') # get there id

                    if '__' in qa_id: # If the id contains '__', it indicates a clause type
                        clause_type = qa_id.split('__')[-1] # Extract the clause type from the id
                        unique_clause_types.add(clause_type) # Add the clause type to the set
    print(f"\nUnique Clause Types Found ({len(unique_clause_types)}):")

    for ct in sorted(unique_clause_types): # sort then print each unique clause type
            print(f"- {ct}")


if __name__ == "__main__": # Run the following code only if this file is executed directly.
    dataset_path = "../CUAD_Dataset/CUAD_v1.json" # dataset path
    inspect_cuad_data(dataset_path) # calling the function to inspect the CUAD dataset
```
